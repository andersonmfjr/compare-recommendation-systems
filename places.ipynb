```python
import pandas as pd
```

```python
data = pd.read_csv('tourism.csv')
```

```python
data.head()
```

```python
item_names_id = data[['item_id', 'item_name']]
```

```python
item_names_id.head()
```

```python
item_names_id.drop_duplicates(inplace=True)
```

```python
item_names_id.sort_values(by=['item_id'], inplace=True)
```

```python
item_names_id.reset_index(inplace=True)
```

```python
item_names_id.drop(['index'], axis=1, inplace=True)
```

```python
item_names_id
```

```python
item_names_id.to_csv('places.csv', index=False)
```
